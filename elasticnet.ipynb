```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('/content/drive/MyDrive/data/baseball.csv')
```

```python
data
```

```python
x=data.drop(columns=['player','salary'])
y=data['salary']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
model_elasticnet=ElasticNet(alpha=0.3, max_iter=10000, tol=0.000001)
```

```python
model_elasticnet.fit(x_train,y_train)
```

```python
model_elasticnet.score(x_train,y_train)
```

```python
model_elasticnet.score(x_test,y_test)
```
